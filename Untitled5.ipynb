```python
import pandas as pd
```

```python
students=pd.read_csv('/CompleteDataAndBiases.csv')
students.head(10)
```

```python
students.tail(10)
```

```python
students.dtypes
```

```python
students=students.drop(['freetime','romantic','G3'],axis=1)
students.head(10)
```

```python
students=students.rename(columns={"sex":"gender","StereotypeActivation":"activation"})
students.head(10)
```

```python
students.shape
```

```python
duplicate_rows_students = students[students.duplicated()]
print("number of duplicate rows: ", duplicate_rows_students.shape)
```

```python
students.count()
```

```python
students=students.drop_duplicates()
students.head(10)
```

```python
students.count()
```

```python
print(students.isnull().sum())
```

```python
students= students.dropna()
students.count()
```

```python
import matplotlib.pyplot as plt
```

```python
students.name.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title('students analysis')
plt.ylabel('pass')
plt.xlabel('name');
#
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(students['name'], students['Pass'])
ax.set_xlabel('name')
ax.set_ylabel('pass')
plt.show()
```
